# 使用tf.keras.datasets加载数据

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
mnist = tf.keras.datasets.mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape,  y_test.shape)
```

```python
y_train = y_train.astype('float32')
y_test = y_test.astype('float32')
```

```python
# Reserve 10,000 samples for validation

x_val = x_train[-10000:]
y_val = y_train[-10000:]
x_train = x_train[:-10000]
y_train = y_train[:-10000]
```

```python
print(x_train.shape)
print(x_val.shape)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
fig = plt.figure()

for i in range(15):
    plt.subplot(3, 5, i+1)    # 绘制前15个手写体数字，以3行5列子图形式展示
    plt.tight_layout()        # 自动适配子图尺寸
    plt.imshow(x_train[i], cmap='Greys')        # 使用灰色显示像素灰度值
    plt.title('Label: {}'.format(y_train[i]))   # 设置标签为子图标题
    plt.xticks([])    # 删除x轴标记
    plt.yticks([])    # 删阶y轴标签
```

# 使用 tf.keras 管理Sequential模型

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
tf.keras.utils.plot_model(model)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=10)
```

```python
model.evaluate(x_val, y_val)
```

# 保存为SavedModel格式模型

```python
model.save('mnist_model')
```

# 加载SavedModel格式模型

```python
sm = tf.keras.models.load_model('mnist_model')
```

```python
sm.evaluate(x_val, y_val)
```

# 保存为HDF5格式模型

```python
model.save('mnist_model.h5')
```

# 加falgHDF5格式模型

```python
hm = tf.keras.models.load_model('mnist_model.h5')
```

```python
hm.evaluate(x_val, y_val)
```

# 使用 tf.keras 管理 functional API

## 使用tf.keras.Input 定义模型的输入

```python
inputs = tf.keras.Input(shape=(28, 28))
x = tf.keras.layers.Flatten(input_shape=(28, 28))(inputs)
x = tf.keras.layers.Dense(128, activation='relu')(x)
x = tf.keras.layers.Dropout(0.2)(x)
outputs = tf.keras.layers.Dense(10, activation='softmax')(x)
```

```python
model = tf.keras.Model(inputs=inputs, outputs=outputs, name='mnist_model')
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=10)
```

```python
model.evaluate(x_val, y_val)
```
